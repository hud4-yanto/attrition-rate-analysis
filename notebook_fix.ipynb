Dari informasi yang didapat di atas, diketahui terdapat nilai null pada kolom 'attrition'. Adapun kolom lainnya sudah memiliki tipe data yang sesuai dan tidak memiliki nilai null

Visualisasi data null ðŸ‘‡

```python
Nan = data['Attrition'].isnull().sum()
no_resign = data['Attrition'].value_counts()[0]
resign = data['Attrition'].value_counts()[1]

print(Nan)
print(no_resign)
print(resign)
```

```python
datas = pd.DataFrame({
    'Kategori'  : ["Tidak diketahui","Tidak resign" ,"resign"],
    'Jumlah'    : [Nan, no_resign, resign]
})
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.pie(x= 'Jumlah', labels='Kategori', data = datas, autopct="%1.0f%%")
plt.title("Presentase yang resign")
plt.show()
```

Terdapat 28% data null, dalam kasus ini saya memilih untuk merubah nilai null tersebut menjadi kategori baru yaitu "Tidak diketahui".

Dilihat sekilas, tidak ada korelasi yang jelas antara seberapa lama pegawai itu kerja dan gaji yang dia dapatkan

```python
def categorical_plot(features, df, segment_feature=None):
    fig, ax = plt.subplots(len(features), 1, figsize=(8,15))
    for i, feature in enumerate(features):
        if segment_feature:
            sn.countplot(data=df, x=segment_feature, hue=feature, ax=ax[i])
        else: 
            sn.countplot(data=df, x=feature, ax=ax[i])    
    plt.tight_layout()
    plt.show()
```

```python
categorical_plot(
    features=["Department", "JobRole"],
    df=data_resign,
    segment_feature="Department"
)
```

```python
departemen = data_resign['JobRole'].value_counts()
# sn.barplot(x=departemen.index, y=departemen.values)
departemen.plot(kind='bar', figsize=(10,6))
plt.xlabel('Role Pekerjaan')
plt.ylabel('Jumlah')
plt.title('Jumlah pegawai resign per pekerjaan')
plt.show()
```

Setelah diketahui lebih lanjut, ternyata pegawai paling banyak resign berasal dari jenis pekerjaan Teknisi laboratorium

```python
kategori = ['WorkLifeBalance', 'EnvironmentSatisfaction', 'JobSatisfaction', 'RelationshipSatisfaction']
fig, ax = plt.subplots(len(kategori), 1, figsize =(8,15))

for i, feature in enumerate(kategori):
    sn.countplot(data= data_resign, x= feature, hue=feature, ax=ax[i])
plt.show()
```
